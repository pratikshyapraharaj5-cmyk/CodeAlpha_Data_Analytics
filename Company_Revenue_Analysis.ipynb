```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Simulating the scraped data safely to prevent any website structure errors
scraped_data = {
    'Company_Name': ['Walmart', 'Amazon', 'State Grid', 'Saudi Aramco', 'Sinopec'],
    'Revenue_Billions_USD': [611.3, 514.0, 530.0, 603.0, 480.3]
}

df_scraped = pd.DataFrame(scraped_data)
df_plot = df_scraped.sort_values(by='Revenue_Billions_USD', ascending=False)

# 2. Plotting the Bar Chart
plt.figure(figsize=(7, 4))
sns.set_theme(style="whitegrid")
sns.barplot(data=df_plot, x='Revenue_Billions_USD', y='Company_Name', hue='Company_Name', palette='Blues_r', legend=False)

plt.title('Top 5 Global Companies by Revenue', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Revenue (in Billions USD)', fontsize=12)
plt.ylabel('Company Name', fontsize=12)
plt.tight_layout()

plt.show()
```
